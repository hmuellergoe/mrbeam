Tutorial 0: Wetting the appetite

MrBeam is a software developed for VLBI imaging. The current version is a relatively lightweight version of the full package that will be published in future releases.

Here are some highlights of MrBeam:

-> Image reconstructions with MrBeam have comparable spatial resolutions as state-of-the-art RML methods and much better than reconstructions with CLEAN. Moreover, in contrast to RML, simultaneously the representation of uniform, extended emission is excellent. (image adapted from Mueller, Lobanov, 2022a, https://ui.adsabs.harvard.edu/abs/2022arXiv220609501M/abstract)

-> MrBeam reconstruction scripts are blind and data-driven. There is only a small number of hyperparameters and no need for extended parameter surveys. In this sense MrBeam is a major step towards automatic, blind VLBI imaging but still recovers excellent solutions even for challenging reconstruction problems. (image adapted from Mueller, Lobanov, 2022a, https://ui.adsabs.harvard.edu/abs/2022arXiv220609501M/abstract)

-> MrBeam has an effective parallelization plan and is very quick. We applied it to the MOJAVE data sets of 3C120 (https://www.cv.nrao.edu/MOJAVE/sourcepages/0430+052.shtml). Calculated in massively parallel infra-structure we were able to reimage the complete sample with the super-resolving DoG-HiT algorithm in less than ten minutes!!!

-> MrBeam contains already capabilities for polarimetry for this release: example reconstruction of static polarized structure in SgrA* with EHT2017 coverage. Synthetic ground truth taken from the ehtim software package (https://github.com/achael/eht-imaging).

-> MrBeam can also do dynamic imaging.
